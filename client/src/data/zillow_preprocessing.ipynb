### Drop home values with unknown zipcodes

```python
drop_indices = home_values[~home_values['RegionName'].isin(zips['zip'])].index

home_values.drop(drop_indices, inplace=True)

home_values.reset_index(drop=True, inplace=True)

home_values.info()
```

### Drop rentals with unknown zipcodes

```python
drop_indices = rentals[~rentals['RegionName'].isin(zips['zip'])].index

rentals.drop(drop_indices, inplace=True)

rentals.reset_index(drop=True, inplace=True)

rentals.info()
```
